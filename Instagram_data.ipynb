```python
df = df.drop(['Hashtags'], axis = 1)
```

```python
df.head()
```

```python
x = df.iloc[:, 1:]
x.head()
```

```python
y = df.iloc[:, :1]
y.head()
```

```python
from sklearn.model_selection import train_test_split as tts
```

```python
xtrain,xtest,ytrain,ytest = tts(x, y, test_size = .2, random_state = 42)
```

```python
from sklearn.linear_model import LinearRegression
lrn = LinearRegression()
lrn.fit(xtrain,ytrain)
```

```python
lrn.score(xtest, ytest)
```
